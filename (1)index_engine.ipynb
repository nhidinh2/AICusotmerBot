```python
import subprocess
def run_command(command):
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    if result.returncode != 0:
        print(f"Error running command: {command}")
        print(result.stderr)
    else:
        print(result.stdout)

# Step 1: Install pipx
run_command("python -m pip install --user pipx")
```

```python

# Step 2: Install Poetry using pipx
run_command("python -m pipx install poetry")
```

```python
run_command("python -m pip install pkginfo==1.10.0")
```

```python

# Step 4: Install dependencies using Poetry
run_command("python -m poetry install")
```
