```python
import subprocess

neo4j_bin_path = '/Users/dinhuyennhi/untitled folder/neo4j-community-5.21.2/bin'

def start_neo4j():
    start_command = [f'{neo4j_bin_path}/neo4j', 'start']
    subprocess.run(start_command)
    print("Neo4j started.")

def stop_neo4j():
    stop_command = [f'{neo4j_bin_path}/neo4j', 'stop']
    subprocess.run(stop_command)
    print("Neo4j stopped.")

def check_status():
    status_command = [f'{neo4j_bin_path}/neo4j', 'status']
    status_result = subprocess.run(status_command, capture_output=True, text=True)
    print("Neo4j status:\n", status_result.stdout)
```

```python
# Start Neo4j
start_neo4j()

# Check Neo4j status
check_status()

# Stop Neo4j
#stop_neo4j()

```
